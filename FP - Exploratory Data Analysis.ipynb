## Exploratory Data Analysis

> Tujuan : Menganalisis konsumen Amazon. Dari gender, usia, dan produk yang dibeli. Dan menganalisis kenapa amazon diminati oleh konsumen dan bisa bersaing dengan perusahaan multi nasional lainnya.

```python
from sklearn.preprocessing import LabelEncoder

data = amazon_db

# Buat salinan DataFrame
data_encoded = data.copy()

# Lakukan encoding untuk setiap kolom kategorikal
for column in data_encoded.columns:
    if data_encoded[column].dtype == object:
        lbl = LabelEncoder()
        lbl.fit(list(data_encoded[column].values))
        data_encoded[column] = lbl.transform(list(data_encoded[column].values))

# Sekarang semua fitur adalah numerik, kita bisa membuat heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(data_encoded.corr(), annot=True, cmap='coolwarm')

plt.title('Heatmap dari Database')
plt.show()
```

> Prinsip Heatmap

- Semakin positif, nilai antara feature berbanding lurus
- Semakin negatif, nilai antara feature berbanding negatif
- Semakin mendekati 0, tidak ada korelasi

```python
def get_random_color():
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)

color = get_random_color()

plt.figure(figsize = (20,8))
plt.hist(data=amazon_db,x='age', color=color)
plt.title('Konsumen Amazon Berdasarkan Umur')
plt.show()
```

Berdasarkan grafik tersebut dapat disimpulkan bahwa individu pada rentang usia 20-25 tahun memiliki tingkat interaksi pengguna yang lebih tinggi dibandingkan kelompok usia lainnya. Hal ini mengisyaratkan bahwa individu dalam rentang usia ini lebih banyak terlibat berpartisipasi.

```python
!pip install squarify
import matplotlib.pyplot as plt
import squarify

# Misalkan 'amazon_db' adalah DataFrame Anda dan 'Browsing_Frequency' adalah kolom yang Anda inginkan
data = amazon_db['Browsing_Frequency']

# Hitung jumlah nilai unik
value_counts = data.value_counts()

# Normalisasi data untuk ukuran persegi panjang
sizes = value_counts.values / value_counts.sum()

squarify.plot(sizes=sizes, label=value_counts.index, alpha=0.6)
plt.axis('off')

plt.title('Seberapa Sering Konsumen Mengakses Web/App Amazon')
plt.show()
```

> 'Beberapa kali seminggu', 'Beberapa kali sebulan', 'Jarang', 'Beberapa kali sehari'

Data menunjukan Konsumen mengakses web/aplikasi amazon beberapa kali dalam seminggu. Yang menunjukan bahwa amazon berhasil menjadi salah satu perusahaan teknologi multinasional favorit. Dengan Tampilan web/aplikasi yang nyaman dan mudah digunakan.

> 'Kadang-kadang', 'Sangat', 'Sedang', 'Tidak Pernah', 'Jarang'

Dari Histogram data diatas. Diambil dari data Ulasan/Review dan Manfaat review. Histogram menunjukan bahwa semakin sering terjadinya review, semakin banyak konsumen yang melakukan review produk amazon, Maka Semakin bermanfaat pula review tersebut dikarenakan adanya informasi tambahan mengenai produk dan layanan amazon.

```python
data = amazon_db['Improvement_Areas']

# Hitung jumlah nilai unik dan ambil 5 teratas
top_5 = data.value_counts().nlargest(4)

# Buat salinan data
data_copy = data.copy()

# Ubah nilai yang bukan termasuk dalam 5 teratas menjadi 'others'
data_copy[~data_copy.isin(top_5.index)] = 'others'

# Hitung jumlah nilai unik
value_counts = data_copy.value_counts()

# Buat label dan ukuran dari data
labels = value_counts.index
sizes = value_counts.values

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Komentar Terhadap Amazon')

plt.show()
```

Data menunjukkan, Komentar yang paling banyak tentang amazon yaitu :
1. Responsif pelayanan pelanggan (Costumer service responsiveness) = 36%
2. Kualitas dan keakuratan produk (Product quality and accuracy) = 26% 
3. Mengurangi sampah kemasan (Reducing packaging waste) = 22%
4. Kecepatan dan keandalan pengiriman (Shipping speed and reliability) = 13%
